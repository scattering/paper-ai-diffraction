# Reviewer Walkthrough: Single-Pattern Extinction-Group Inference

This notebook is the reviewer-facing inference path for the paper repo. It accepts a plain CSV with `2theta` and `intensity`, normalizes and interpolates the pattern onto the model grid, runs extinction-group inference from a released checkpoint, and optionally compares the prediction to a supplied true space group on the condensed extinction-group DAG.

The notebook also ships with two real `RRUFF-325` examples:

- one correct case under the final mixed paper model
- one failure case under the same model


## Setup

Before running this notebook, install the train/eval environment and editable package:

```bash
conda env create -f environment-train-eval.yml
conda activate paper-ai-diffraction-train-eval
pip install -e .
```

Place the released checkpoint under `external/checkpoints/` as documented in `reproducibility/checkpoint_manifest.csv`.

If you want paper-faithful calibrated inference, also provide a `PRIOR_H5` path. Otherwise the notebook still works in checkpoint-only mode.


````python
from pathlib import Path
import json
import matplotlib.pyplot as plt
import pandas as pd

from paper_ai_diffraction.reviewer import (
    build_model_bundle,
    describe_topology_relation,
    infer_single_pattern,
    load_example_manifest,
    load_example_metadata,
    load_precomputed_benchmark_summary,
    load_topology_assets,
    normalize_and_interpolate_pattern,
    read_pattern_csv,
    space_group_to_ext_group,
)
````

````python
cwd = Path.cwd().resolve()
REPO_ROOT = next((candidate for candidate in [cwd, *cwd.parents] if (candidate / 'configs').exists()), None)
if REPO_ROOT is None:
    raise FileNotFoundError(f'Could not locate repo root from {cwd}')
CHECKPOINT_PATH = REPO_ROOT / 'external/checkpoints/xrd_model_82ept35h_best.pth'
CONFIG_PATH = REPO_ROOT / 'configs/final_mixed_2500k_dualsource.json'
PRECOMPUTED_SUMMARY_PATH = REPO_ROOT / 'results/reviewer/rruff325_precomputed_inference.json'
PRIOR_H5 = None  # optionally set to a trainready HDF5 for paper-faithful calibrated inference
DEVICE = None    # set to 'cpu' or 'cuda' if you want to force a device
DECODER = 'aux_bayes'  # falls back automatically if PRIOR_H5 is not supplied
AUX_TEMPERATURE = 5.0

bundle = build_model_bundle(
    checkpoint_path=CHECKPOINT_PATH,
    config_path=CONFIG_PATH,
    device=DEVICE,
    prior_h5_path=PRIOR_H5,
)
topology = load_topology_assets()
bundle.config['two_theta_min'], bundle.config['two_theta_max'], bundle.config['spec_length']
````

## Select A Pattern CSV

You can either point `PATTERN_CSV` at one of the bundled real examples or replace it with your own CSV. The shipped manifest and paired metadata files let you cross-check the paper-backed correct and failure examples without touching HDF5 inputs.


````python
example_dir = REPO_ROOT / 'assets/reviewer_examples'
example_manifest = load_example_manifest(example_dir / 'manifest.json')
example_manifest
````

````python
example_csvs = sorted(example_dir.glob('*case_*.csv'))
PATTERN_CSV = next(example_dir.glob('correct_case_*.csv'))
# PATTERN_CSV = next(example_dir.glob('failure_case_*.csv'))
# PATTERN_CSV = Path('/path/to/your_pattern.csv')
selected_metadata = load_example_metadata(PATTERN_CSV) if PATTERN_CSV.parent == example_dir and PATTERN_CSV.with_suffix('.json').exists() else None
PATTERN_CSV
````

````python
df = read_pattern_csv(PATTERN_CSV)
prep = normalize_and_interpolate_pattern(
    df,
    two_theta_min=float(bundle.config.get('two_theta_min', 5.0)),
    two_theta_max=float(bundle.config.get('two_theta_max', 90.0)),
    spec_length=int(bundle.config['spec_length']),
)
prep['warnings']
````

````python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(prep['source_two_theta'], prep['source_intensity'], label='source CSV', alpha=0.7)
ax.plot(prep['target_two_theta'], prep['target_intensity'], label='model-grid interpolation', linewidth=1)
ax.set_xlabel('2θ')
ax.set_ylabel('Normalized intensity')
ax.legend()
plt.show()
````

````python
result = infer_single_pattern(
    bundle,
    prep['target_intensity'],
    aux_temperature=AUX_TEMPERATURE,
    decoder=DECODER,
    prior_h5_path=PRIOR_H5,
)
result
````

## Interpret The Prediction

The model predicts an extinction group, not a unique space group. The compatible space groups listed below are the crystallographically valid space groups for that predicted extinction group.


````python
selected = result['decoders'][result['selected_decoder']]
pd.DataFrame({
    'rank': range(1, len(selected['top5_eg']) + 1),
    'predicted_eg': selected['top5_eg'],
    'probability': selected['top5_prob'],
})
````

````python
print('Top-1 extinction group:', result['top1_eg'])
print('Canonical symbol:', result['top1_symbol'])
print('Crystal system:', result['top1_crystal_system'])
print('Compatible space groups:', result['top1_candidate_space_groups'])
````

## Optional Truth-Aware Topology Check

For shipped examples, the notebook uses the paired metadata automatically when the true space group is available. For an arbitrary reviewer-uploaded CSV, you can enter the true space group or true extinction group manually. The notebook then reports whether the prediction is exact, ancestor, descendant, or a branch jump in the condensed DAG used in the paper.


````python
TRUE_SPACE_GROUP = None  # override for an arbitrary uploaded CSV, e.g. 19
TRUE_EG = None           # optional direct override if you know the true extinction group instead

metadata_true_sg = selected_metadata.get('true_space_group') if selected_metadata else None
metadata_true_eg = selected_metadata.get('true_eg') if selected_metadata else None
metadata_true_sg_symbol = selected_metadata.get('true_space_group_symbol') if selected_metadata else None
metadata_true_sg_source = selected_metadata.get('true_space_group_source') if selected_metadata else None

resolved_true_sg = TRUE_SPACE_GROUP if TRUE_SPACE_GROUP is not None else metadata_true_sg
resolved_true_eg = None
truth_source = None
sg_mapped_eg = None

if resolved_true_sg is not None:
    sg_mapped_eg = space_group_to_ext_group(resolved_true_sg, bundle)

if metadata_true_eg is not None:
    resolved_true_eg = int(metadata_true_eg)
    truth_source = 'benchmark_metadata_extinction_group'
elif TRUE_EG is not None:
    resolved_true_eg = int(TRUE_EG)
    truth_source = 'manual_extinction_group'
elif sg_mapped_eg is not None:
    resolved_true_eg = sg_mapped_eg
    truth_source = 'space_group_only'

if resolved_true_eg is not None:
    topo = describe_topology_relation(result['top1_eg'], resolved_true_eg, topology)
    if resolved_true_sg is not None:
        print('True space group:', resolved_true_sg)
        if metadata_true_sg_symbol:
            print('True space group symbol:', metadata_true_sg_symbol)
        if metadata_true_sg_source:
            print('True space group source:', metadata_true_sg_source)
        if sg_mapped_eg is not None:
            print('Space-group-mapped extinction group:', sg_mapped_eg)
    if metadata_true_eg is not None and sg_mapped_eg is not None and int(metadata_true_eg) != int(sg_mapped_eg):
        print('Warning: supplied true space group maps to a different extinction group than the benchmark metadata. Using benchmark true_eg for topology.')
    print('Resolved true extinction group:', resolved_true_eg)
    print('Truth source:', truth_source)
    print('Topology relation:', topo['relation'])
    print('Shortest-path distance:', topo['distance'])
    print('Condensed DAG path:', topo['path'])
else:
    print('Set TRUE_SPACE_GROUP or TRUE_EG to enable DAG comparison for an uploaded CSV.')
````

## Precomputed Paper Benchmark Summary

If `results/reviewer/rruff325_precomputed_inference.json` is present, the notebook can browse the frozen 325-example paper summary directly. This is the fast path for TAP: reviewers can inspect benchmark-backed predictions, candidate space groups, and topology locality without rerunning batch inference.


````python
precomputed = load_precomputed_benchmark_summary(PRECOMPUTED_SUMMARY_PATH, bundle) if PRECOMPUTED_SUMMARY_PATH.exists() else None
if precomputed is None:
    print('Precomputed summary not found at', PRECOMPUTED_SUMMARY_PATH)
else:
    print('Loaded', precomputed['n_examples'], 'examples from', PRECOMPUTED_SUMMARY_PATH)
    pd.DataFrame([
        {
            'index': ex['index'],
            'case_id': ex['case_id'],
            'mineral': ex.get('mineral'),
            'true_eg': ex.get('true_eg'),
            'top1_eg': ex.get('top1_eg'),
            'top1_probability': ex.get('top1_probability'),
            'decoder': ex.get('selected_decoder'),
        }
        for ex in precomputed['examples'][:10]
    ])
````

````python
PRECOMPUTED_INDEX = 0
if precomputed is not None:
    ex = precomputed['examples'][PRECOMPUTED_INDEX]
    print('Case ID:', ex['case_id'])
    print('Mineral:', ex.get('mineral'))
    print('True extinction group:', ex.get('true_eg'))
    print('Predicted extinction group:', ex.get('top1_eg'))
    print('Top-1 probability:', ex.get('top1_probability'))
    print('Candidate space groups:', ex.get('top1_candidate_space_groups'))
    print('Top-5 extinction groups:', ex.get('top5_eg'))
    if ex.get('true_eg') is not None and ex.get('top1_eg') is not None:
        topo = describe_topology_relation(ex['top1_eg'], ex['true_eg'], topology)
        print('Topology relation:', topo['relation'])
        print('Shortest-path distance:', topo['distance'])
        print('Condensed DAG path:', topo['path'])
else:
    print('No precomputed summary loaded.')
````
